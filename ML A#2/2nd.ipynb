## Single Input Multiple Weight Single Output
Add Learning Rate in the code (take help from 1st Notebook)

```python
from tqdm import tqdm
```

```python
def forward_pass(data, w1, w2):
    prediction1 = data*w1
    prediction2 = data*w2
    prediction = prediction1 + prediction2
    return round(prediction,2)
    
def calculate_error(actual_output, prediction):
    error = actual_output - prediction
    return round(error,2)

def backward_pass(error, w1, w2):
    if error > 0:
        w1 = w1 + 0.2
        w2 = w2 + 0.2
    elif error < 0:
        w1 = w1 - 0.2
        w2 = w2 - 0.2
        
    return round(w1,2),round(w2,2)
    
```

```python
data = 5
actual_output = 10
w1 = 0.5
w2 = 0.1
```

```python
for i in tqdm(range(10)):
    prediction = forward_pass(data=data, w1=w1, w2=w2)
    error = calculate_error(actual_output=actual_output, prediction=prediction)
    w1, w2 = backward_pass(error=error, w1=w1, w2=w2)
    print("Prediction : ", prediction)
    print("Error : ", error)
    print("Updated Weight : ", w1, w2)
    print("==============================")
```

```python
import matplotlib.pyplot as plt

data = 5
actual_output = 10
w1 = 0.5
w2 = 0.1
predictions = []
errors = []
w1_updates = []
w2_updates = []

for i in tqdm(range(10)):
    prediction = forward_pass(data=data, w1=w1, w2=w2)
    error = calculate_error(actual_output=actual_output, prediction=prediction)
    w1, w2 = backward_pass(error=error, w1=w1, w2=w2)
    predictions.append(prediction)
    errors.append(error)
    w1_updates.append(w1)
    w2_updates.append(w2)

# Plot the predictions and errors
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(predictions, 'bo-', label='Prediction')
plt.plot([actual_output]*len(predictions), 'r--', label='Actual')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.legend()

plt.subplot(1,2,2)
plt.plot(errors, 'ro-')
plt.xlabel('Iteration')
plt.ylabel('Error')

# Plot the weight updates
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(w1_updates, 'bo-')
plt.xlabel('Iteration')
plt.ylabel('w1')

plt.subplot(1,2,2)
plt.plot(w2_updates, 'ro-')
plt.xlabel('Iteration')
plt.ylabel('w2')

plt.show()
```
